```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
re = pd.read_csv('https://raw.githubusercontent.com/zd2425/bootcamp_final/refs/heads/main/real_estate_dataset.csv')
```

```python
re.head()
```

```python
re.info()
```

```python
re["Price"].describe()
```

```python
# Boxplot of prices
sns.boxplot(data = re, x = "Price")
plt.title("Boxplot of House Prices")
```

```python
# Histogram of Prices
sns.histplot(data = re, x = "Price")
plt.title("Histogram of House Prices")
```

```python
# Price vs Bedrooms
sns.barplot(data = re, x = "Num_Bedrooms", y = "Price")
plt.title("Average Price by Number of Bedrooms")
```

```python
# Price vs Bathrooms
sns.barplot(data = re, x = "Num_Bathrooms", y = "Price")
plt.title("Average Price by Number of Bathrooms")
```

```python
# Price vs Floors
sns.barplot(data = re, x = "Num_Floors", y = "Price")
plt.title("Average Price by Number of Floors")
```

```python
# Price vs Area
sns.regplot(data = re, y = "Price", x = "Square_Feet")
plt.title("Price by Square Feet")
```

```python
# Price vs Year Built
sns.regplot(data = re, y = "Price", x = "Year_Built")
plt.title("Price by Year Built")
```

```python
# Price vs Pool
sns.barplot(data = re, x = "Has_Pool", y = "Price")
plt.title("Average Price by Pool")
```

```python
# Price vs Garden
sns.barplot(data = re, x = "Has_Garden", y = "Price")
plt.title("Average Price by Garden")
```

```python
# Price vs Garage Size
sns.regplot(data = re, x = "Garage_Size", y = "Price")
plt.title("Price by Garage Size")
```

```python
# Price vs Location Score
sns.regplot(data = re, y = "Price", x = "Location_Score")
plt.title("Price by Location Score")
```

```python
# Price vs Distance to Center
sns.regplot(data = re, y = "Price", x = "Distance_to_Center")
plt.title("Price by Distance to Center")
```

```python
# Compute correlation matrix
correlation_matrix = re.corr()

# Extract correlations of features with the target variable (Price)
correlation_with_price = correlation_matrix['Price'].sort_values(ascending=False)

# Create heatmap for all features including Price
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Feature Correlation Heatmap')
plt.show()
```
